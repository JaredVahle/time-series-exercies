**Our dataframe is now downsized into the mean average temp for each year starting with 1781 and ending 2013 this spans 233 years of data and will be what we use moving forward in exploration**

```python
train = df['1781-01-01':'1920-01-01']
validate = df['1920-01-02':'1970-01-01']
test = df['1970-01-02':]
```

```python
len(train) + len(validate) + len(test) == len(df)
```

```python
# lets plot our split
ax = train.avg_temp.plot(label = 'train')
validate.avg_temp.plot(ax = ax, label = 'validate')
test.avg_temp.plot(ax=ax, label = 'test')
plt.legend()
plt.show()
```

```python
# lets continue exploring our train dataset
ax = train.avg_temp.plot(label='yearly average', alpha=.7)
train.avg_temp.rolling(5).mean().plot(ax=ax, label='moving average 5 years',alpha = .7)
plt.title("yearly avg rolling 5 years")
ax.legend()
plt.show()
```

```python
ax = train.avg_temp.plot(label='yearly', alpha=.7)
train.avg_temp.rolling(10).mean().plot(ax=ax, label='moving average 5 years',alpha = .7)
plt.title('yearly avg rolling 10 years')
ax.legend()
plt.show()
```

```python
# lets look at a scatterplot of year over year change
pd.concat([df.avg_temp.rename('t'), df.avg_temp.shift().rename('t + 1')], axis=1).plot.scatter(y='t + 1', x='t')
plt.title('Year over year change')
plt.ylabel('temp of next year')
plt.xlabel('temp of current year')
plt.show()
```

**We can see some linearity when looking at the year over year change in temperature**

```python
# making a lag plot for 2 years
pd.plotting.lag_plot(train.avg_temp, 2)
plt.title('temp lag plot of 2 years')
plt.ylabel('temp at year + 2')
plt.ylabel('temp at current year')
plt.show()
```

```python
# now lets look at 3
pd.plotting.lag_plot(train.avg_temp, 3)
plt.title('temp lag plot of 3 years')
plt.ylabel('temp at year + 3')
plt.ylabel('temp at current year')
plt.show()
```

```python
# on to 4 we can see linearity decreasing each year
pd.plotting.lag_plot(train.avg_temp, 4)
plt.title('temp lag plot of 4 years')
plt.ylabel('temp at year + 4')
plt.ylabel('temp at current year')
plt.show()
```

```python
# lets try 5 at this point linearity is compleatly gone
pd.plotting.lag_plot(train.avg_temp, 5)
plt.title('temp lag plot of 5 years')
plt.ylabel('temp at year + 5')
plt.ylabel('temp at current year')
plt.show()
```

**When looking at the lag plot of 5 years that linearity is gone there is some clustering in the center though**

```python
def autocorrelation(train, lag):
    return pd.concat([train.avg_temp, train.avg_temp.shift(lag)], axis=1).dropna().corr().iloc[0, 1]
```

```python
# this will cycle through the lag to see if there is any reoccuring correlations in the data
autocorr = pd.Series({lag: autocorrelation(train, lag) for lag in range(1, 100)})
autocorr.plot()
```

```python
# Lets plot the 20 year mark its very interesting that linearity increases could the
# temperature follow some longer trends, or is this complete coincidence?
pd.plotting.lag_plot(train.avg_temp, 20)
plt.title('temp lag plot of 5 years')
plt.ylabel('temp at year + 5')
plt.ylabel('temp at current year')
plt.show()
```

```python
# since we are using year over year data we didnt expect to find much seasonality in the data
# This still shows that the correlation has downward trends with spikes upward
```

```python
pd.plotting.autocorrelation_plot(train.avg_temp)
plt.title('lag autocorrelation of temp')
plt.show()
```

**Takeaways**
- year over year change seems to stay somewhat consistent you wont have huge swings in average temperatures yearly.
- We saw that there would be spans of years when the average temperature would drop followed by a steep increase in temperature that resets the cycle.
- We also found that looking at the data with a 5 year lag has almost no correlation, meaning that over longer spans of time it is hard to predict the temperature given the temperature of many years before.
- Even more interesting we can find our data cycling every 20-30 years of lag although this could all be coincidence, **No huge assumtions should be made on one singular dataset**
