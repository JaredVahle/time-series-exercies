**When looking at the lag plot of 5 years that linearity is gone there is some clustering in the center though**

```python
def autocorrelation(train, lag):
    return pd.concat([train.avg_temp, train.avg_temp.shift(lag)], axis=1).dropna().corr().iloc[0, 1]
```

```python
# this will cycle through the lag to see if there is any reoccuring correlations in the data
autocorr = pd.Series({lag: autocorrelation(train, lag) for lag in range(1, 100)})
autocorr.plot()
plt.title('Correlation')
plt.xlabel('years')
plt.ylabel('correlation percentage')
plt.grid()
plt.show()
```

```python
# Lets plot the 20 year mark its very interesting that linearity increases could the
# temperature follow some longer trends, or is this complete coincidence?
pd.plotting.lag_plot(train.avg_temp, 20)
plt.title('temp lag plot of 5 years')
plt.ylabel('temp at year + 5')
plt.ylabel('temp at current year')
plt.grid()
plt.show()
```

```python
# since we are using year over year data we didnt expect to find much seasonality in the data
# This still shows that the correlation has downward trends with spikes upward
```

```python
pd.plotting.autocorrelation_plot(train.avg_temp)
plt.title('lag autocorrelation of temp')
plt.show()
```

**Takeaways**
- year over year change seems to stay somewhat consistent you wont have huge swings in average temperatures yearly.
- We saw that there would be spans of years when the average temperature would drop followed by a steep increase in temperature that resets the cycle.
- We also found that looking at the data with a 5 year lag has almost no correlation, meaning that over longer spans of time it is hard to predict the temperature given the temperature of many years before.
- Even more interesting we can find our data cycling every 20-30 years of lag although this could all be coincidence, **No huge assumtions should be made on one singular dataset**

# holt's linear trend

```python
model = Holt(train['avg_temp'], exponential = False)
# fit the model providing the alpha/ smoothing level and beta/smoothing slope
model = model.fit(smoothing_level = .4,
                     smoothing_slope = .16,
                     optimized = False)

 # predict/forecast providing the start and end dates
yhat = model.predict(start = validate.index[0], end = validate.index[-1])

yhat_df['avg_temp'] = round(yhat, 2)
```

```python
# plotting the predictions
plot_and_eval(train,validate,test,yhat_df,target_var = 'avg_temp')
plt.grid()
plt.title("holt's linear trend predictions")
plt.xlabel('year')
plt.ylabel('temp')
plt.show()
```

```python
# adding holts linear model predictions to a dataframe for the RMSE
eval_df = append_eval_df(eval_df,validate,yhat_df,model_type = ("holt's linear trend"),target_var = "avg_temp")
```

```python
# Checking RMSE
eval_df
```

# Previous cycle

```python
# made a function to find the optimal cycle
def find_best_cycle():
    rmse_tracker = 3
    for i in range(1,70):
        yhat_df = train['1870':] + train.diff(i).mean()
        yhat_df = yhat_df.set_index(validate.index)
        rmse = evaluate(validate,yhat_df,target_var = 'avg_temp')
        if rmse < rmse_tracker:
            rmse_tracker = rmse
            year_tracker = i
    return rmse_tracker, year_tracker
```

```python
# looking for the best cycle for RMSE
find_best_cycle()
```

```python
# The best cycle is 44 years
```

```python
# make predictions
yhat_df = train['1870':] + train.diff(44).mean()
yhat_df = yhat_df.set_index(validate.index)
```

```python
# plotting the predictions
plot_and_eval(train,validate,test,yhat_df,target_var = 'avg_temp')
plt.grid()
plt.title('previous cycle predictions')
plt.xlabel('year')
plt.ylabel('temp')
plt.show()
```

```python
# adding previous cycle to a dataframe for the RMSE
eval_df = append_eval_df(eval_df,validate,yhat_df,model_type = ("previous cycle_44_years"),target_var = "avg_temp")
```

```python
# Checking RMSE
eval_df
```
